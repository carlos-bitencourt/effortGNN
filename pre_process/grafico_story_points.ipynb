```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter1d

# Ler dados
df_sp = pd.read_csv('issues_sp_filtered.csv')

# Filtrar Story Points at 100
df_filtrado = df_sp[df_sp['Story_Point'] <= 100].copy()
contagem_sp = df_filtrado['Story_Point'].value_counts().sort_index()

story_points = contagem_sp.index.values
quantidades = contagem_sp.values


# Grfico de barras
plt.figure(figsize=(14, 6))
plt.bar(story_points, quantidades, label='Frequncia')


# Aparncia
plt.xlabel('Story Points')
plt.ylabel('Number of Issues (Frequency)')
plt.title('Distribution of Story Points (up to 100 points)')
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.tight_layout()
plt.savefig('graficos/distribuicao_story_points.png', dpi=300)
plt.show()
```
